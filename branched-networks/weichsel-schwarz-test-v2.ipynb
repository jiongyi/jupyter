```python
# Create actin network object and evolve over time.
from weichselSchwarz2 import *
from numpy import array
from numpy.random import normal

n = network(1, 7410 * 30.0/20.0, 18.5, 2, 5.7)
n.evolve(2.7e-3, 20.0)
```

```python
# Plot nucleation over time.
from numpy import linspace, gradient, mean
from matplotlib.pyplot import plot, xlabel, ylabel
%matplotlib inline

noBranchesArr = array([b for a, b, c in n.noFilaments])
timeArr = linspace(0, n.tElapsed, len(noBranchesArr))
plot(timeArr, noBranchesArr)
xlabel(r"Time (s)", fontsize = 18)
ylabel(r"Branches (/s)", fontsize = 18)
```

```python
# Plot network growth over time.
plot(timeArr, n.xEdge)
xlabel(r"Time (s)", fontsize = 18)
ylabel(r"Leading-edge position (nm)", fontsize = 18)
```

```python
noCapsArr = array([c for a, b, c in n.noFilaments])
plot(timeArr, noCapsArr)
xlabel(r"Time (s)", fontsize = 18)
ylabel(r"Caps", fontsize = 18)
```

```python
# Plot kernel density estimate of the distribution of orientations of the active filaments.
from scipy import stats

kde = stats.gaussian_kde(n.allThetaArr)
x = linspace(-pi, pi, 100)
p = kde(x)
plot(x / pi * 180, p)
xlabel(r"Angle", fontsize = 18)
ylabel(r"Estimated PDF", fontsize = 18)
```
