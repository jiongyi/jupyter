# Rate equations

\begin{align}
\frac{d[E]}{dt} &= k_N [E] [WA]^2 - k_C [E] \\
\frac{d[WA]}{dt} &= (k_A + k_T \frac{[PA]}{[WA]_{max}}) ([WA]_{max} - [WA]) - k_N [E][WA]^2 - k_P [E][WA] \\
\frac{d[PA]}{dt} &= (k_{PA} - k_T \frac{[PA]}{[WA]_{max}})([WA]_{max} - [WA]) - k_P [E][PA]
\end{align}

At steady state $\frac{d[E]}{dt} = 0$, $\frac{d[WA]}{dt} = 0$, and $\frac{d[PA]}{dt} = 0$:

\begin{align}
[WA]_{eq} &= \sqrt{\frac{k_C}{k_N}} \\
[E]_{eq} &= \frac{(k_A + k_T \frac{[PA]_{eq}}{[WA]_{max}})([WA]_{max} - \sqrt{\frac{k_C}{k_N}})}{k_P \sqrt{\frac{k_C}{k_N}}} \\
[PA]_{eq} &= \frac{\sqrt{k_T^2 \frac{k_C}{k_N} + k_A^2 [WA]_{max}^2 + 2 k_T (k_A + 2 k_{PA})[WA]_{max} \sqrt{\frac{k_C}{k_N}}} - (k_T \sqrt{\frac{k_C}{k_N}} + k_A[WA]_{max})}{2 k_T}
\end{align}

```python
from math import sqrt
from numpy import zeros, array, linspace
from matplotlib.pyplot import plot, xlabel, ylabel
%matplotlib inline

def nucleationmodel(kP = 100.0, kN = 10.0, kC = 0.1, kA = 10.0, kPA = 10.0, kT = 0.1):
    noPoints = 20000
    WAMax = 1.0
    dt = 1e-3
    EArr = zeros(noPoints)
    WAArr = zeros(noPoints)
    PAArr = zeros(noPoints)
    timeArr = zeros(noPoints)
    EArr[0] = 0.1
    
    for i in range(noPoints - 1):
        E = EArr[i]
        WA = WAArr[i]
        PA = PAArr[i]
        dEdt = kN * E * WA**2 - kC * E
        dWAdt = (kA + kT * PA / WAMax) * (WAMax - WA) - kN * E * WA**2 - kP * E * WA
        dPAdt = (kPA - kT * PA / WAMax) * (WAMax - WA) - kP * E * PA
        EArr[i + 1] = EArr[i] + dEdt * dt
        WAArr[i + 1] = WAArr[i] + dWAdt * dt
        PAArr[i + 1] = PAArr[i] + dPAdt * dt
        timeArr[i + 1] = timeArr[i] + dt
        
    #Plot.
    #plot(timeArr, EArr)
    #print(str(EArr[-1]))
        
    # Figure out steady-state quantities.
    #WAEq = sqrt(kC / kN)
    #PAEq = sqrt((kT / WAMax + kA / WAEq)**2 + 4 * kT * kPA / WAMax / WAEq) - (kT / WAMax + kA / WAEq)
    #PAEq = PAEq / (2 * kT / WAMax / WAEq)
    #EEq = (kA + kT * PAEq / WAMax) * (WAMax - WAEq) / kP / WAEq
    WAEq = WAArr[-1]
    PAEq = PAArr[-1]
    EEq = EArr[-1]
    
    return EEq, PAEq, WAEq
```

```python
noPoints = 100
capRateArr = linspace(0.01, 10.0, noPoints)
nucleationRateArr = zeros(noPoints)
elongationRateArr = zeros(noPoints)

for i in range(noPoints):
    EEq, PAEq, WAEq = nucleationmodel(100.0, 10.0, capRateArr[i], 10.0, 10.0, 1.0)
    nucleationRateArr[i] = 10.0 * EEq * WAEq**2
    elongationRateArr[i] = 100.0 * EEq * (WAEq + PAEq)
    
plot(capRateArr, nucleationRateArr)
xlabel(r"Capping rate (1/s)", fontsize = 18)
ylabel(r"Nucleation rate", fontsize = 18)
```

```python
plot(capRateArr, elongationRateArr)
xlabel(r"Capping rate (1/s)", fontsize = 18)
ylabel(r"Elongation rate (1/s)", fontsize = 18)
```

```python
noPoints = 100
polRateArr = linspace(0.01, 100.0, noPoints)
nucleationRateArr = zeros(noPoints)
elongationRateArr = zeros(noPoints)

for i in range(noPoints):
    EEq, PAEq, WAEq = nucleationmodel(polRateArr[i], 10.0, 5, 10.0, 10.0, 1.0)
    nucleationRateArr[i] = 10.0 * EEq * WAEq**2
    elongationRateArr[i] = polRateArr[i] * EEq * (WAEq + PAEq)
    
plot(polRateArr, nucleationRateArr)
xlabel(r"Polymerization rate (1/s)", fontsize = 18)
ylabel(r"Nucleation rate", fontsize = 18)
```

```python
noPoints = 100
branchRateArr = linspace(0.01, 100.0, noPoints)
nucleationRateArr = zeros(noPoints)
elongationRateArr = zeros(noPoints)

for i in range(noPoints):
    EEq, PAEq, WAEq = nucleationmodel(100.0, branchRateArr[i], 5, 10.0, 10.0, 1.0)
    nucleationRateArr[i] = branchRateArr[i] * EEq * WAEq**2
    elongationRateArr[i] = 100.0 * EEq * (WAEq + PAEq)
    
plot(branchRateArr, nucleationRateArr)
xlabel(r"Branching rate (1/s)", fontsize = 18)
ylabel(r"Nucleation rate", fontsize = 18)
```
