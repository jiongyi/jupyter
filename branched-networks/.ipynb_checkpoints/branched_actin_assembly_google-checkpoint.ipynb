```python
from numpy import array, zeros, round, linspace
def simulate_assembly(elongationRate = 100.0, nucleationRate = 1.0, 
                      cappingRate = 0.1, pprLoadingRate = 1.0, 
                      transferRate = 0.1):
  # Define numerical simulation constants.
  timeInterval = 1e-3
  totalTime = 25.0
  npfDensity = 2000.0

  # Perform simulation.
  numberIterations = int(round(totalTime / timeInterval))
  timeArr = linspace(0.0, totalTime, numberIterations)
  filamentNumberArr = zeros(numberIterations)
  wh2FilledDensityArr = zeros(numberIterations)
  pprFilledDensityArr = zeros(numberIterations)
  filamentNumberArr[0] = 1.0
  for i in range(numberIterations - 1):
    changeFilamentNumber = (nucleationRate * wh2FilledDensityArr[i]**2 * filamentNumberArr[i] - 
                            cappingRate * filamentNumberArr[i]) * timeInterval
    changePprFilledDensity = (pprLoadingRate * (npfDensity - pprFilledDensityArr[i]) - 
                              transferRate * pprFilledDensityArr[i] * (npfDensity - wh2FilledDensityArr[i]) - 
                              elongationRate * pprFilledDensityArr[i] * filamentNumberArr[i]) * timeInterval
    changeWh2FilledDensity = (transferRate * pprFilledDensityArr[i] * (npfDensity - wh2FilledDensityArr[i]) - 
                             nucleationRate * wh2FilledDensityArr[i]**2 * filamentNumberArr[i] - 
                             elongationRate * wh2FilledDensityArr[i] * filamentNumberArr[i]) * timeInterval
    filamentNumberArr[i + 1] = filamentNumberArr[i] + changeFilamentNumber
    pprFilledDensityArr[i + 1] = pprFilledDensityArr[i] + changePprFilledDensity
    wh2FilledDensityArr[i + 1] = wh2FilledDensityArr[i] + changeWh2FilledDensity
  return filamentNumberArr[-1], pprFilledDensityArr[-1], wh2FilledDensityArr[-1]
```

```python
# Capping-protein dependence
cappingRateArr = linspace(0.1, 10.0, 40)
eqFilamentNumberArr = zeros(40)
eqPprFilledDensityArr = zeros(40)
eqWh2FilledDensityArr = zeros(40)
for i in range(40):
  eqFilamentNumberArr[i], eqPprFilledDensityArr[i], eqWh2FilledDensityArr[i] = simulate_assembly(cappingRate = cappingRateArr[i])
```

```python
%matplotlib inline
from matplotlib import style
from matplotlib.pyplot import legend, subplots, savefig

elongationRateArr = 100.0 * (eqWh2FilledDensityArr + eqPprFilledDensityArr) * eqFilamentNumberArr
nucleationRateArr = 1.0 * eqWh2FilledDensityArr * eqFilamentNumberArr

style.use('seaborn-notebook')
figObj, axObj = subplots(nrows = 1, ncols = 2, sharex=True, squeeze=True)

figObj.set_figheight(3)
figObj.set_figwidth(6)

axObj[0].axhline(linewidth = 1)
axObj[0].axvline(linewidth = 1)
axObj[0].set_ylim([2, 11])
axObj[0].tick_params(direction = 'out')
axObj[0].plot(cappingRateArr, eqFilamentNumberArr, linewidth = 2, label = r'$[E]_{eq}$')
axObj[0].plot(cappingRateArr, nucleationRateArr, linewidth = 2, label = 'nucleation rate')
axObj[0].set_xlabel(r'$k_C$ (/s)', fontsize = 11)
axObj[0].set_ylabel('Steady-state', fontsize = 11)
axObj[0].tick_params(axis = 'both', labelsize = 11)
axObj[0].legend(fontsize = 11)

axObj[1].axhline(linewidth = 1)
axObj[1].axvline(linewidth = 1)
axObj[1].tick_params(direction = 'out')
axObj[1].plot(cappingRateArr, eqPprFilledDensityArr, linewidth = 2, label = r'$[PA]_{eq}$')
axObj[1].plot(cappingRateArr, eqWh2FilledDensityArr, linewidth = 2, label = r'$[WA]_{eq}$')
axObj[1].set_xlabel(r'$k_C$ (/s)', fontsize = 11)
axObj[1].legend(fontsize = 11)
axObj[1].tick_params(axis = 'both', labelsize = 11)

figObj.tight_layout()
```
