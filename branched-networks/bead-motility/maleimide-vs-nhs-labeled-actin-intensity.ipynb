```python
# Load image stack.
from skimage.io import imread, imshow

fileName = '/home/jiongyi/Documents/2018-07-19-200-nM-n-wasp-lcb-50-nM-arp23-5-uM-profilin-ii-malei-actin-ss-gallery' + '/2018-07-19-200-nM-n-wasp-lcb-50-nM-arp23-5-uM-profilin-ii-malei-actin-ss-gallery_MMStack.ome.tif' 
stack = imread(fileName)
```

```python
# Subtract background
noImages, m, n = stack.shape
from numpy import zeros
from skimage.morphology import white_tophat, disk
tophatFilteredImages = zeros((m, n, noImages))

selem = disk(50)
for i in range(noImages):
    tophatFilteredImages[:, :, i] = white_tophat(stack[i, :, :], selem)
    
imshow(tophatFilteredImages[:, :, 1])
```
