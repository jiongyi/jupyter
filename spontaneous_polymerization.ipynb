```python
import numpy as np
import matplotlib.pyplot as plt


# Initialize variables.
kNuc = 1e-2
kPol = 1
N = 10000
dt = 1e-3
t = np.arange(0, N * dt, dt)
F = np.zeros(N)
G = np.zeros(N)
G[0] = 2
n = 3
# Iterate.
for i in range(0, N - 1):
    dFdt = kNuc * G[i]**n + kPol * F[i] * G[i]
    dGdt = - kPol * F[i] * G[i]
    F[i + 1] = F[i] + dFdt * dt
    G[i + 1]= G[i] + dGdt * dt
# Plot.
plt.plot(t, F, t, G)
plt.xlabel('Time (s)')
plt.ylabel('F(t)')
plt.show()

```
