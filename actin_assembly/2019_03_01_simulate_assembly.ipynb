```python
from npf_branched_actin_assembly import *
```

```python
actin_network = network()
```

```python
actin_network.simulate(60.0)
print(actin_network.x_leading_edge)
print(actin_network.maximum_displacement)
```

```python
%matplotlib inline
from matplotlib.pyplot import subplots
from matplotlib.patches import Rectangle
from numpy import array
line_color_row = array(['b' for x in range(actin_network.no_filaments)])
line_color_row[actin_network.is_capped_row] = 'r'

fig_handle, axes_handle = subplots()
for i in range(actin_network.no_filaments):
    axes_handle.plot([actin_network.x_pointed_row[i], actin_network.x_barbed_row[i]],
                    [actin_network.y_pointed_row[i], actin_network.y_barbed_row[i]],
                    line_color_row[i])
axes_handle.add_patch(Rectangle((actin_network.x_leading_edge, 0), 0.02 * actin_network.x_leading_edge, 1000, color = [0.5, 0.5, 0.5]))
axes_handle.set_xlabel('x (nm)')
axes_handle.set_ylabel('L (nm)')
```
