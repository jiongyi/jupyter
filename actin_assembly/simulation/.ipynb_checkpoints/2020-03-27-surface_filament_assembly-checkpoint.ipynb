```python
from numpy import array, zeros, ones, sin, cos, pi, sqrt, concatenate, mod, sign, abs, exp, sum, min, copy, logical_and
from numpy.random import rand, randn
```

```python
# Define network class.
class network(object):
    def __init__(self):
        self.elongate_rate = 50.0
        self.elongate_force_rate = copy(self.elongate_rate)
        self.branch_rate = 5000.0
        self.cap_rate = 0.0
        self.cap_force_rate = copy(self.cap_rate)
        self.tether_rate = 1.0
        self.no_init_filaments = 200
        
        # Declare constants.
        self.square_length = 1.0 # in microns
        self.monomer_length = 2.7 * 10**-3 # in microns.
        self.mu_theta = 70 / 180 * pi # mean angle at branches, in radians
        self.mu_sigma = 5 / 180 * pi # variance of angle at branches, in radians
        self.bond_stiffness = 1.0
        self.total_simulation_time = 60.0 # in seconds
        self.time_interval = 1e-3 # in seconds
        self.current_time = 0.0 # in seconds

        # Initialize filaments.
        self.x_barb_row = self.square_length * (rand(self.no_init_filaments) - 0.5)
        self.y_barb_row = self.square_length * (rand(self.no_init_filaments) - 0.5)
        self.z_barb_row = self.monomer_length * rand(self.no_init_filaments)

        azi_angle_row = 2 * pi * rand(self.no_init_filaments)
        polar_angle_row = 0.5 * pi * (1 + rand(self.no_init_filaments))
        self.ux_barb_row = sin(polar_angle_row) * cos(azi_angle_row)
        self.uy_barb_row = sin(polar_angle_row) * sin(azi_angle_row)
        self.uz_barb_row = cos(polar_angle_row)
        self.no_barb_ends = self.ux_barb_row.size
        
        self.is_capped_row = zeros(self.no_init_filaments, dtype = bool)
        self.is_tethered_row = zeros(self.no_init_filaments, dtype = bool)
        self.tether_force_row = zeros(self.no_init_filaments)
        self.branch_force_rate_row = self.branch_rate * ones(self.no_init_filaments)
        
    def elongate(self, index):
        self.x_barb_row[index] += self.monomer_length * self.ux_barb_row[index]
        self.y_barb_row[index] += self.monomer_length * self.uy_barb_row[index]
        self.z_barb_row[index] += self.monomer_length * self.uz_barb_row[index]
        # Enforce periodic boundary conditions.
        if abs(self.x_barb_row[index]) > 0.5 * self.square_length:
            self.x_barb_row[index] -= sign(self.x_barb_row[index]) * self.square_length
        if abs(self.y_barb_row[index]) > 0.5 * self.square_length:
            self.y_barb_row[index] -= sign(self.y_barb_row[index]) * self.square_length
                 
    def branch(self, index):
        def rotation_angle_axis(ux_axis, uy_axis, uz_axis, theta_axis):
            R_11 = cos(theta_axis) + ux_axis**2 * (1 - cos(theta_axis))
            R_12 = ux_axis * uy_axis * (1 - cos(theta_axis)) - uz_axis * sin(theta_axis)
            R_13 = ux_axis * uz_axis * (1 - cos(theta_axis)) + uy_axis * sin(theta_axis)
            R_21 = uy_axis * ux_axis * (1 - cos(theta_axis)) + uz_axis * sin(theta_axis)
            R_22 = cos(theta_axis) + uy_axis**2 * (1 - cos(theta_axis))
            R_23 = uy_axis * uz_axis * (1 - cos(theta_axis)) - ux_axis * sin(theta_axis)
            R_31 = uz_axis * ux_axis * (1 - cos(theta_axis)) - uy_axis * sin(theta_axis)
            R_32 = uz_axis * uy_axis * (1 - cos(theta_axis)) + ux_axis * sin(theta_axis)
            R_33 = cos(theta_axis) + uz_axis**2 * (1 - cos(theta_axis))
            rotation_mat = array([[R_11, R_12, R_13], [R_21, R_22, R_23], [R_31, R_32, R_33]])
            return rotation_mat
        
        ux_old = self.ux_barb_row[index]
        uy_old = self.uy_barb_row[index]
        uz_old = self.uz_barb_row[index]
        
        # Find an axis perpendicular to orientation of barbed end.
        u_perp_mag = sqrt(2 + (ux_old + uy_old)**2)
        ux_perp_old = 1.0 / u_perp_mag
        uy_perp_old = 1.0 / u_perp_mag
        uz_perp_old = -(ux_old + uy_old) / u_perp_mag
        
        # Perform rotation to find new orientation.
        theta_polar = self.mu_theta + self.mu_sigma * randn()
        theta_azi = 2 * pi * rand()
        polar_rotation_mat = rotation_angle_axis(ux_perp_old, uy_perp_old, uz_perp_old, theta_polar)
        u_new_polar_row = polar_rotation_mat @ array([ux_old, uy_old, uz_old])
        azi_rotation_mat = rotation_angle_axis(ux_old, uy_old, uz_old, theta_azi)
        u_new_row = azi_rotation_mat @ u_new_polar_row
        
        # Do it until it's facing the right way (-z).
        while u_new_row[2] >= 0.0:
            theta_polar = self.mu_theta + self.mu_sigma * randn()
            theta_azi = 2 * pi * rand()
            polar_rotation_mat = rotation_angle_axis(ux_perp_old, uy_perp_old, uz_perp_old, theta_polar)
            u_new_polar_row = polar_rotation_mat @ array([ux_old, uy_old, uz_old])
            azi_rotation_mat = rotation_angle_axis(ux_old, uy_old, uz_old, theta_azi)
            u_new_row = azi_rotation_mat @ u_new_polar_row
                        
        # Add new barbed end to relevant arrays.
        self.x_barb_row = concatenate((self.x_barb_row, array([self.x_barb_row[index]])))
        self.y_barb_row = concatenate((self.y_barb_row, array([self.y_barb_row[index]])))
        self.z_barb_row = concatenate((self.z_barb_row, array([self.z_barb_row[index]])))
        self.ux_barb_row = concatenate((self.ux_barb_row, array([u_new_row[0]])))
        self.uy_barb_row = concatenate((self.uy_barb_row, array([u_new_row[1]])))
        self.uz_barb_row = concatenate((self.uz_barb_row, array([u_new_row[2]])))
        self.is_capped_row = concatenate((self.is_capped_row, array([False])))
        self.is_tethered_row = concatenate((self.is_tethered_row, array([False])))
        self.tether_force_row = concatenate((self.tether_force_row, array([0])))
        self.no_barb_ends += 1
        
        # Untether end.
        self.is_tethered_row[index] = False
        self.tether_force_row[index] = 0.0
    
    def cap(self, index):
        self.is_capped_row[index] = True
    
    def tether(self, index):
        self.is_tethered_row[index] = True
        self.tether_force_row[index] = self.bond_stiffness * self.z_barb_row[index]
        self.branch_force_rate_row[index] = self.branch_rate * exp(self.tether_force_row[index] / 10.0)
                
    def update_mechanics(self):
        total_tether_force = sum(self.tether_force_row)
        if total_tether_force == 0.0:
            self.elongate_force_rate = copy(self.elongate_rate)
            self.cap_force_rate = copy(self.cap_rate)
        else:
            no_active = sum(logical_and(self.z_barb_row <= self.monomer_length, ~self.is_tethered_row))
            if no_active > 1:
                mean_active_force = total_tether_force / no_active
                force_weight = exp(-mean_active_force * self.monomer_length / 4.114)
                self.elongate_force_rate = self.elongate_rate * force_weight
                self.cap_force_rate = self.cap_rate * force_weight
                
    def update_step(self):
        for index in range(self.no_barb_ends):
            if self.z_barb_row[index] <= self.monomer_length:
                if self.is_capped_row[index] == True:
                    continue
                elif self.is_tethered_row[index] == True:
                    if exp(self.branch_force_rate_row[index] * self.time_interval) < rand():
                        self.branch(index)
                        continue
                elif exp(-self.tether_rate * self.time_interval) < rand():
                    self.tether(index)
                    continue
                elif exp(-self.cap_force_rate * self.time_interval) < rand():
                    self.cap(index)
                    continue
                elif exp(-self.elongate_force_rate * self.time_interval) < rand():
                    self.elongate(index)
            elif self.is_capped_row[index] == True:
                continue
            elif exp(-self.cap_rate * self.time_interval) < rand():
                self.cap(index)
                continue
            elif exp(-self.elongate_rate * self.time_interval) < rand():
                self.elongate(index)
        # Update space, time, and mechanics..
        self.current_time += self.time_interval
        self.z_barb_row -= min(self.z_barb_row)
        self.update_mechanics()
    
    def simulate(self):
        while (self.current_time <= self.total_simulation_time) and (sum(~self.is_capped_row) >= 1):
            self.update_step()
```

```python
test_network = network()
test_network.simulate()
```
