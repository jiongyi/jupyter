```python
from mueller_sixt import Network
from matplotlib.pyplot import subplots
from matplotlib.patches import Rectangle
from numpy import cos, sin
```

```python
test_network = Network()
test_network.catalog_filaments()
```

```python
fig1_hand, axes1_hand = subplots()
axes1_hand.quiver(test_network.barbed_position_mat[:, 0], 
                  test_network.barbed_position_mat[:, 1], 
                  cos(test_network.filament_orientation_row), 
                  sin(test_network.filament_orientation_row), 
                  color='b')
axes1_hand.tick_params(labelsize=14)
axes1_hand.set_xlabel('Monomer-lengths', fontsize=14)
fig1_hand.savefig('figures/initial_network.png', dpi=600)
```

```python
test_network.simulate()
```

```python
fig2_hand, axes2_hand = subplots()
axes2_hand.quiver(test_network.barbed_position_mat[:, 0], 
                  test_network.barbed_position_mat[:, 1], 
                  cos(test_network.filament_orientation_row), 
                  sin(test_network.filament_orientation_row), 
                  color='b')
axes2_hand.tick_params(labelsize=14)
axes2_hand.set_xlabel('Monomer-lengths', fontsize=14)
#axes2_hand.set_xlim(6700, test_network.leading_edge_position)
fig2_hand.savefig('figures/final_network.png', dpi=600)
```

```python
fig3_hand, axes3_hand = test_network.plot_orientation()
axes3_hand.tick_params(labelsize=14)
axes3_hand.set_xlabel('Filament orientation', fontsize=14)
axes3_hand.set_ylabel('Count', fontsize=14)
fig3_hand.savefig('figures/hist_filament_orientation.png', dpi=600)
```
