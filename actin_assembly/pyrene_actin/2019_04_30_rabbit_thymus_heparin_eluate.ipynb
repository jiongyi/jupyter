```python
%matplotlib inline
from pandas import read_csv
from matplotlib.pyplot import subplots
```

```python
pyrene_actin_df = read_csv('/home/jiongyi/Documents/pyrene_actin/2019_04_30_pyrene_actin_rabbit_thymus_heparin_eluate_pm_gst_wasp_wca.csv')
```

```python
from numpy import arange
no_points = len(pyrene_actin_df['Kinetic read'])
time_row = arange(0, 10 * no_points, 10)
fig_handle, axes_handle = subplots()
axes_handle.plot(time_row, pyrene_actin_df['H1'], label = '0 nM WCA')
axes_handle.plot(time_row, pyrene_actin_df['H2'], label = '230 nM WCA')
axes_handle.legend()
axes_handle.set_xlabel('Time (s)', fontsize = 12)
axes_handle.set_ylabel('Pyrene actin (RFU)', fontsize = 12)
axes_handle.tick_params(labelsize = 12)
```
