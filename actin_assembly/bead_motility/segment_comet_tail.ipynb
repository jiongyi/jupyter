```python
file_path_str = '/home/jiongyi/Documents/bead_motility/2019_03_07/4900_nM_profilin_a568_actin_fresh_50_nM_arp23/100_nM_n_wasp_200_nM_fresh_cp/470_gfp_550_a568_wf_50/470_gfp_550_a568_wf_50_MMStack.ome.tif'
```

```python
from skimage.io import imread
from skimage import img_as_float
actin_im = img_as_float(imread(file_path_str)[1, :, :])
```

```python
from skimage.filters import gaussian, threshold_otsu
actin_gauss_im = gaussian(actin_im, sigma = 2)
actin_bw_im = actin_gauss_im > threshold_otsu(actin_gauss_im)
```

```python
from scipy.ndimage import distance_transform_edt
actin_dist_im = distance_transform_edt(actin_bw_im)
```

```python
%matplotlib inline
from matplotlib.pyplot import subplots
fig_hand, axes_hand = subplots()
fig_hand.set_figwidth(20)
fig_hand.set_figheight(20)
dist_threshold = 2.5 / 0.16
print(dist_threshold)
axes_hand.imshow(actin_dist_im)
```
