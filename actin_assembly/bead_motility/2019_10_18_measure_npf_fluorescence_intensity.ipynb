```python
%matplotlib inline
from matplotlib.pyplot import subplots
cp_row = array([25, 50, 75, 100])
wave_mean_tail_growth_rate_row = array([mean(0.5 * cp25_wave_comet_props_mat[:, 2]),
                                        mean(0.5 * cp50_wave_comet_props_mat[:, 2]),
                                        mean(cp75_wave_comet_props_mat[:, 2]),
                                        mean(cp100_wave_comet_props_mat[:, 2])])
wave_std_tail_growth_rate_row = array([std(0.5 * cp25_wave_comet_props_mat[:, 2]),
                                        std(0.5 * cp50_wave_comet_props_mat[:, 2]),
                                        std(cp75_wave_comet_props_mat[:, 2]),
                                        std(cp100_wave_comet_props_mat[:, 2])])

n_wasp_mean_tail_growth_rate_row = array([mean(cp100_n_wasp_props_mat[:, 2]),
                                         mean(cp200_n_wasp_props_mat[:, 2])])
n_wasp_std_tail_growth_rate_row = array([std(cp100_n_wasp_props_mat[:, 2]),
                                        std(cp200_n_wasp_props_mat[:, 2])])

fig1_hand, axes1_hand = subplots()
fig1_hand.set_figheight(5)
fig1_hand.set_figwidth(8)
axes1_hand.errorbar(cp_row, wave_mean_tail_growth_rate_row, yerr = wave_std_tail_growth_rate_row, fmt = 'o-', linewidth = 3, label = 'WAVE')
axes1_hand.errorbar(array([100, 200]), n_wasp_mean_tail_growth_rate_row, yerr = n_wasp_std_tail_growth_rate_row, fmt = 'o-', linewidth = 3, label = 'N-WASP')
axes1_hand.tick_params(labelsize = 14)
axes1_hand.set_xlabel('[CP] (nM)', fontsize = 14)
axes1_hand.set_ylabel('Growth rate ($\mu$m/min)', fontsize = 14)
axes1_hand.legend(fontsize = 14)
```
