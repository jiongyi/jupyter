```python
mm_stack_file_path_list = find_mmstack_files(data_folder_path)
```

```python
mm_stack = imread(mm_stack_file_path_list[5])
```

```python
actin_im = mm_stack[1, :, :]
```

```python
%matplotlib inline
from matplotlib.pyplot import subplots
fig1_hand, axes1_hand = subplots()
axes1_hand.imshow(actin_im, cmap='gray')
```

```python
from skimage.restoration import denoise_tv_chambolle
denoised_im = denoise_tv_chambolle(actin_im, weight = 0.001)
```

```python
from skimage.filters import threshold_minimum
fig2_hand, axes2_hand = subplots()
bw_im = denoised_im > threshold_minimum(denoised_im)
axes2_hand.imshow(bw_im)
```

```python
from skimage.measure import regionprops, label
bw_properties = regionprops(label(bw_im),intensity_image = denoised_im)
```

```python
from skimage.io import imshow
imshow(bw_properties[0].intensity_image, cmap='gray')
```
