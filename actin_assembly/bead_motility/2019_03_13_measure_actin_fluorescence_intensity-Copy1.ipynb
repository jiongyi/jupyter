# Actin density on N-WASP- and WAVE-coated beads

```python
from bead_motility import *
n_wasp_200_data_folder_path_str = '/home/jiongyi/Documents/bead_motility/2019_03_13/2_hr_100_nM_n_wasp_pwwca_5_uM_profilin_a568_actin_50_nM_arp23_200_nM_cp'
wave_200_data_folder_path_str = '/home/jiongyi/Documents/bead_motility/2019_03_13/2_hr_100_nM_wave_pwca_5_uM_profilin_a568_actin_50_nM_arp23_200_nM_cp'
```

```python
n_wasp_200_cp_npf_au_row, n_wasp_200_cp_actin_au_row = batch_measure_actin_density(n_wasp_200_data_folder_path_str, save_images = False)
wave_200_cp_npf_au_row, wave_200_cp_actin_au_row = batch_measure_actin_density(wave_200_data_folder_path_str, save_images = False)
```

```python
%matplotlib inline
from matplotlib.pyplot import figure
from seaborn import swarmplot, boxplot
from pandas import DataFrame
normal_actin_intensity_df = DataFrame.from_dict(data = {'N-WASP 200 nM CP': n_wasp_200_cp_actin_au_row / n_wasp_200_cp_npf_au_row,
                                    'WAVE 200 nM CP': wave_200_cp_actin_au_row / wave_200_cp_npf_au_row}, orient = 'index')

npf_intensity_df = DataFrame.from_dict(data = {'N-WASP 200 nM CP': n_wasp_200_cp_npf_au_row,
                                    'WAVE 200 nM CP': wave_200_cp_npf_au_row}, orient = 'index')



fig_handle = figure()
fig_handle.set_figheight(8)
fig_handle.set_figwidth(10)
box_handle = boxplot(data = npf_intensity_df.transpose())
swarm_handle = swarmplot(data = npf_intensity_df.transpose(), color = "0.2")
swarm_handle.set_ylabel('NPF AU', fontsize = 16)
swarm_handle.tick_params(labelsize=16)
```

```python
fig_handle = figure()
fig_handle.set_figheight(8)
fig_handle.set_figwidth(10)
box_handle = boxplot(data = normal_actin_intensity_df.transpose())
swarm_handle = swarmplot(data = normal_actin_intensity_df.transpose(), color = "0.2")
swarm_handle.set_ylabel('Actin AU / NPF AU', fontsize = 16)
swarm_handle.tick_params(labelsize=16)
```
