```python
# Import.
import trackpy as tp
from skimage.io import imread, imshow
from numpy import mean, gradient, sqrt, array, nanmean, nanstd, cumsum, ones
from matplotlib.pyplot import subplots
```

```python
# Load stack.
mm_stack = imread('/home/jiongyi/Documents/bead_motility/470_egfp_npf_570_a568_actin_z_time_2_MMStack.ome.tif')
```

```python
# Z-project and extract EGFP channel.
z_stack = mean(mm_stack, axis = 2)
z_egfp_stack = z_stack[:, 0, :, :]
```

```python
# Locate and link features.
features_df = tp.batch(z_egfp_stack[:], 5, minmass = 1000)
trajectories_df = tp.link_df(features_df, 8, memory = 3)
filtered_trajectories_df = tp.filter_stubs(trajectories_df, 3)
drift_df = tp.compute_drift(trajectories_df)
corrected_df = tp.subtract_drift(filtered_trajectories_df, drift_df)
trajectories_df.head()
```

```python
%matplotlib inline
tp.mass_size(corrected_df.groupby('particle').mean())
```

```python
axes_hand = tp.plot_traj(corrected_df, superimpose = mean(z_egfp_stack, axis = 0))
```

```python
# Extract xy data from line objects.
no_trajectories = len(axes_hand.lines)
fig_hand, axes2_hand = subplots()
for i in range(no_trajectories):
    i_x_row = axes_hand.lines[i].get_xdata()
    i_y_row = axes_hand.lines[i].get_xdata()
    i_time_row = cumsum(ones(len(i_x_row)) / 3)
    i_dx_row = gradient(i_x_row)
    i_dy_row = gradient(i_y_row)
    i_dr_row = sqrt(i_dx_row**2 + i_dy_row**2)
    axes2_hand.plot(i_time_row, i_dr_row * 3)
    print([nanmean(i_dr_row * 3), nanstd(i_dr_row * 3)])
axes2_hand.set_xlabel('Time (min)', fontsize = 12)
axes2_hand.set_ylabel('Speed ($\mu$m / min)', fontsize = 12)
```
