```python
# Process truncation (pwca)
delta25_folder_path = '/home/jiongyi/Documents/lab_data/2020_08_27/egfp_n_wasp_pwca_lcbs/25_nM_by_20min/'
delta75_folder_path = '/home/jiongyi/Documents/lab_data/2020_08_27/egfp_n_wasp_pwca_lcbs/75_nM_by_20min'
delta100_folder_path = '/home/jiongyi/Documents/lab_data/2020_08_27/egfp_n_wasp_pwca_lcbs/100_nM_by_20min_redo/'
delta150_folder_path = '/home/jiongyi/Documents/lab_data/2020_08_27/egfp_n_wasp_pwca_lcbs/150_nM_by_20min/'
delta300_folder_path = '/home/jiongyi/Documents/lab_data/2020_08_27/egfp_n_wasp_pwca_lcbs/300_nM_by_20min/'

delta25_props_mat = batch_analysis(delta25_folder_path)
delta75_props_mat = batch_analysis(delta75_folder_path)
delta100_props_mat = batch_analysis(delta100_folder_path)
delta150_props_mat = batch_analysis(delta150_folder_path)
delta300_props_mat = batch_analysis(delta300_folder_path)
```

```python
%matplotlib inline
from matplotlib.pyplot import subplots
from numpy import nanmean, nanstd
wt_mean_intensity_row = array([nanmean(wt25_props_mat), 
                               nanmean(wt75_props_mat), 
                               nanmean(wt100_props_mat), 
                               nanmean(wt150_props_mat),
                               nanmean(wt300_props_mat)])

wt_std_intensity_row = array([nanstd(wt25_props_mat), 
                               nanstd(wt75_props_mat), 
                               nanstd(wt100_props_mat), 
                               nanstd(wt150_props_mat),
                               nanstd(wt300_props_mat)])

delta_mean_intensity_row = array([nanmean(delta25_props_mat), 
                               nanmean(delta75_props_mat), 
                               nanmean(delta100_props_mat), 
                               nanmean(delta150_props_mat),
                               nanmean(delta300_props_mat)])

delta_std_intensity_row = array([nanstd(delta25_props_mat), 
                               nanstd(delta75_props_mat), 
                               nanstd(delta100_props_mat), 
                               nanstd(delta150_props_mat),
                               nanstd(delta300_props_mat)])


fig1_hand, axes1_hand = subplots()
fig1_hand.set_figheight(5)
fig1_hand.set_figwidth(8)
npf_conc_row = [25, 75, 100, 150, 300]
axes1_hand.errorbar(npf_conc_row, wt_mean_intensity_row, yerr = wt_std_intensity_row, fmt = 'o-', linewidth = 3, label = 'N-WASP PWWCA')
axes1_hand.errorbar(npf_conc_row, delta_mean_intensity_row, delta_std_intensity_row, fmt = 'o-', linewidth = 3, label = 'N-WASP PWCA')
axes1_hand.tick_params(labelsize = 14)
axes1_hand.set_xlabel('[NPF] (nM)', fontsize = 14)
axes1_hand.set_ylabel('Mean EGFP intensity (AU)', fontsize = 14)
axes1_hand.legend(fontsize = 14)
```
