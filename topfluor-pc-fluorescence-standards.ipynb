# TopFluor-PC/EGFP fluorescence standards

```python
from numpy import array, zeros
from scipy import stats
%matplotlib inline
from matplotlib.pyplot import plot, xlabel, ylabel

dataMat = array([[0.2298411452, 12731.514], [0.111345544, 8291.856], [0.0548200815,4462.729], [0.02720172837, 2694.628]])
numberLipidsPerMicron = 1 / 0.72 / 1e-6
dataMat[:, 0] *= numberLipidsPerMicron / 1e2
slope, intercept, r_value, p_value, std_err = stats.linregress(dataMat[:, 0],dataMat[:, 1])
print(r_value)
plot(dataMat[:, 0], dataMat[:, 1], 'b.')
plot(dataMat[:, 0], slope * dataMat[:, 0] + intercept, 'r')
xlabel("TopFluor-PC (/$\mu m^2$)", fontsize = 18)
ylabel("Average pixel intensity", fontsize = 18)
```

```python
solDataMat = array([[0.111345544 * 30.84257169 * 1e1, 9617.44], [0.0548200815 * 31.32230771 * 1e1, 5545.455], [0.02720172837 * 31.56217572 * 1e1, 3280.156]])
slope, intercept, r_value, p_value, std_error = stats.linregress(solDataMat[:, 0], solDataMat[:, 1])
print(r_value)
plot(solDataMat[:, 0], solDataMat[:, 1], 'b.')
plot(solDataMat[:, 0], slope * solDataMat[:, 0] + intercept, 'r')
xlabel("TopFluor-PC ($\mu M$)", fontsize = 18)
ylabel("Average pixel intensity in sol.", fontsize = 18)
intensityTopFluor = slope
```

```python
egfpDataMat = array([[3.064404432, 6813.983], [1.532202216, 4594.704], [0.766101108, 4285.364], [0.383050554, 3667.287]])
slope, intercept, r_value, p_value, std_error = stats.linregress(egfpDataMat[:, 0], egfpDataMat[:, 1])
print(r_value)
plot(egfpDataMat[:, 0], egfpDataMat[:, 1], 'b.')
plot(egfpDataMat[:, 0], slope * egfpDataMat[:, 0] + intercept, 'r')
xlabel("EGFP ($\mu M$)", fontsize = 18)
ylabel("Average pixel intensity in sol.", fontsize = 18)
```

```python
egfp2DataMat = array([[3.064404432, 10568.813],  [1.532202216, 4273.751], [0.766101108, 1699.327], [0.383050554, 812.394]])
slope, intercept, r_value, p_value, std_error = stats.linregress(egfp2DataMat[:, 0], egfp2DataMat[:, 1])
print(r_value)
plot(egfp2DataMat[:, 0], egfp2DataMat[:, 1], 'b.')
plot(egfp2DataMat[:, 0], slope * egfp2DataMat[:, 0] + intercept, 'r')
xlabel("EGFP ($\mu M$)", fontsize = 18)
ylabel("Average pixel intensity in sol.", fontsize = 18)
print(slope)
intensityEGFP = slope
```
