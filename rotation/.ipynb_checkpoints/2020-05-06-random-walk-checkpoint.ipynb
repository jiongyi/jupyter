```python
from numpy import arange, zeros, hstack
from matplotlib.pyplot import subplots
```

```python
def compute_probabilities(no_steps):
    position_row = arange(-no_steps, no_steps + 1)
    probability_row = zeros(position_row.size)
    position_outcome_row = 0
    msd_row = zeros(no_steps)
    for i_step in range(1, no_steps):
        position_outcome_row = hstack((position_outcome_row - 1, position_outcome_row + 1))
        msd_row[i_step] = (position_outcome_row**2).mean()
    for i_position in range(position_row.size):
        probability_row[i_position] = (position_outcome_row == position_row[i_position]).sum()
    probability_row /= probability_row.sum()
    return position_row, probability_row, msd_row
```

```python
fig1_hand, axes1_hand = subplots(ncols=2, nrows=1)
fig1_hand.set_figwidth(9)
x_row, p_row, msd_row = compute_probabilities(11)
axes1_hand[0].bar(x_row, p_row)
axes1_hand[0].set_xlabel('Position')
axes1_hand[0].set_ylabel('Probability')
axes1_hand[1].plot(arange(11), msd_row)
axes1_hand[1].set_xlabel('No of moves')
axes1_hand[1].set_ylabel('Mean squared displacement')
```
