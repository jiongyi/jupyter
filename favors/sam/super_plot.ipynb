```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel

#import dataset called "pooled" that has the columns "Replicate," "Treatment," and "Speed." Change these and the code below if necessary.
pooled = pd.read_csv("pooled.csv")

sns.set(style="whitegrid")

ReplicateAverages = pooled.groupby(['Treatment','Replicate'], as_index=False).agg({'Speed': "mean"})

#calculate P value using paired t test
ReplicateAvePivot = ReplicateAverages.pivot_table(columns='Treatment', values='Speed', index="Replicate")
statistic, pvalue = ttest_rel(ReplicateAvePivot['Control'], ReplicateAvePivot['Drug'])

P_value = str(float(round(pvalue, 3)))

sns.swarmplot(x="Treatment", y="Speed", hue="Replicate", data=pooled)
ax = sns.swarmplot(x="Treatment", y="Speed", hue="Replicate", size=15, edgecolor="k", linewidth=2, data=ReplicateAverages)
ax.legend_.remove()
x1, x2 = 0, 1   # columns 'Control' and 'Drug'
y, h, col = pooled['Speed'].max() + 2, 2, 'k'
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
plt.text((x1+x2)*.5, y+h*2, "P = "+P_value, ha='center', va='bottom', color=col)
```
